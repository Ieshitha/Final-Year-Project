```python
import pandas as pd  #Model creation for soundRepition
df = pd.read_csv('sep28k-mfcc.csv')
```

```python
df = df[df['NoStutteredWords'] != 0]
df = df[df['NaturalPause'] == 0]
df = df[df['Interjection'] == 0]
df = df[df['Prolongation'] == 0]
df = df[df['WordRep'] == 0]
df = df[df['Block'] == 0]
df
```

```python
df.loc[df['SoundRep'] >= 1.0, 'SoundRep'] = 1.0
df['SoundRep'].value_counts()
```

```python
X = df.iloc[: , -13:]
X
```

```python
y = df['SoundRep']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
from collections import Counter
Counter(y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train , y_train)
knn.score(X_test , y_test) * 100
```

```python
import numpy as np
y_pred = np.array(knn.predict(X_test))
y_actual = np.array(y_test)
print(Counter(y_pred))
print(Counter(y_actual))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
clf = RandomForestClassifier(n_estimators = 1000)
# clf.fit(X_over, y_over)
# clf.fit(X_smote, y_smote)
clf.fit(X_train, y_train)
# clf.fit(X_pipe, y_pipe)
y_pred = clf.predict(X_test)
metrics.accuracy_score(y_test, y_pred) * 100
```

```python
y_pred = np.array(clf.predict(X_test))
y_actual = np.array(y_test)
print(Counter(y_pred))
print(Counter(y_actual))
```

```python
from sklearn.neural_network import MLPClassifier
nn_model = MLPClassifier(random_state=1, max_iter=300).fit(X_train, y_train)
nn_model.score(X_test, y_test) * 100
```

```python
y_pred = np.array(nn_model.predict(X_test))
y_actual = np.array(y_test)
print(Counter(y_pred))
print(Counter(y_actual))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='entropy',random_state=5)
clf.fit(X_train, y_train)
clf.score(X_test,y_test) * 100
```

```python
y_pred = np.array(clf.predict(X_test))
y_actual = np.array(y_test)
print(Counter(y_pred))
print(Counter(y_actual))
```
