```python
import os 
CLIPS_DIR = "OneDrive/Documents/SEP_28k/clips/stuttering-clips/clips/"
ignore_list = []
for filename in os.listdir(CLIPS_DIR):
    file_path = CLIPS_DIR + filename
    if 'FluencyBank' in filename:
        if os.stat(file_path).st_size == 44:
            ignore_list.append(filename)
            filename = filename[:-4]
            df = df[df.FileName != filename]

print(len(ignore_list))
# get all the empty audio files into a list and ignore them
```

```python
#MFCC feature Extraction
import librosa
import numpy as np
from tqdm.notebook import tqdm

features = {}
directory = CLIPS_DIR

for filename in tqdm(os.listdir(CLIPS_DIR)):
    filename = filename[:-4]
    if 'FluencyBank' in filename and ignore_list.count(filename + '.wav') == 0:
        audio, sample_rate = librosa.load(CLIPS_DIR + filename + '.wav', res_type='kaiser_fast', sr=None)
        mfccs = np.mean(librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=13).T,axis=0)
        features[filename] = mfccs
```

```python
print(len(features))
```

```python
#make dataset from features
df_features = pd.DataFrame.from_dict(features)
df_features = df_features.transpose()
df_features
```

```python
df_features = df_features.reset_index()
df_features
```

```python
df_features = df_features.sort_values(by='index')
df_features
```

```python
df_features['index']
```

```python
#Apply inner join on the dataframes
df_features.rename(columns ={'index': 'FileName'} , inplace = True)
df_features
```

```python
df_final = pd.merge(df , df_features , how='inner' , on='FileName')
df_final
```

```python
df_final.head()
```

```python
df_final.info()
```

```python
#ignore audio with poor quality
df_final = df_final[df_final.PoorAudioQuality == 0]
df_final
```

```python
#ignore the audios which is difficult to understand
df_final = df_final[df_final.DifficultToUnderstand == 0]
df_final
```

```python
#ignore music audio files
df_final = df_final[df_final.Music == 0]
df_final
```

```python
#ignore audios which does not contain any speech
df_final = df_final[df_final.NoSpeech == 0]
df_final
```

```python
#finally get the cleaned data into a csv file
df_final.to_csv("fluency-mfcc.csv",index=False)
```
